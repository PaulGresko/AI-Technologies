```python
# Удаление строк с пропущенными значениями
df_cleaned = df.dropna()
```

```python
# Заполнение пропусков средним
df_filled = df.copy()
df_filled['salary_from'] = df_filled['salary_from'].fillna(df_filled['salary_from'].mean())
```

```python
# Новое поле через выражение
df['avg_salary'] = (df['salary_from'] + df['salary_to']) / 2
```

```python
# Новое поле через apply(DataFrame)
df['salary_range'] = df[['salary_from', 'salary_to']].apply(
    lambda row: row['salary_to'] - row['salary_from'] if pd.notnull(row['salary_from']) and pd.notnull(row['salary_to']) else None, axis=1)
```

```python
# Новое поле через apply(Series)
df['experience_level'] = df['experience'].apply(lambda x: x.split()[0] if pd.notnull(x) else 'Unknown')
```

```python
# Сортировка
df_sorted = df.sort_values(by='avg_salary', ascending=False)
```

```python
# Статистики
df[['salary_from', 'salary_to']].agg(['mean', 'max', 'min', 'std'])
```

```python
# value_counts
df['experience'].value_counts()
```

```python
# unique
df['schedule'].unique()
```

```python
# Новый индекс
df = df.reset_index(drop=True)
df = df.set_index('published_at')
```

### 4. Продемонстрировать работу `.groupby`, на основе группировок в `groupby` вычислить агрегатные функции по одной или нескольким колонкам.


```python
df.groupby('experience')['avg_salary'].mean()
```

```python
df.groupby(['experience', 'employment'])['avg_salary'].agg(['mean', 'count'])
```

### 5. Решейпинг данных $1D\rightarrow2D$ с использованием `.pivot` (можно подать на вход результаты агрегатов, полученных ранее с использованием `.groupby` (сгруппировать по двум полям).


```python
pivot_df = df.groupby(['experience', 'employment'])['avg_salary'].mean().reset_index()
pivot_df.pivot(index='experience', columns='employment', values='avg_salary')
```

### 11. Продемонстировать работу `.merge`.
